## Merging Weather and Tweets data frames

```python
import pandas as pd
import numpy as np
```

* We read the two previously saved csv dataframes

```python
tweets = pd.read_csv('Tweets_analysed.csv', low_memory=False)
wlabels = pd.read_csv('weatherlabels.csv', low_memory=False)
dailyweather = pd.read_csv('dailyweather.csv')
```

* We check them

```python
wlabels.head()
dailyweather.head()
wlabels['Temperature'].value_counts()
```

```python
tweets = tweets.sort_values('tweet_dates')
tweets.head(3)
```

* We know we have some problems with some columns where the values moved and the dates are not correct anymore. We are going to delete these rows.
* Also, we will delete some unnecessary columns.

```python
tweets = tweets.drop(columns=['tweet_id', 'tweet_name', 'OBJECTID', 'Unnamed: 0'])
wlabels = wlabels.drop(columns=['Unnamed: 0'])
dailyweather = dailyweather.drop(columns=['Unnamed: 0', 'MinTemp', 'MaxTemp', 'MinPres', 'MaxPres', 'MinHum', 'MaxHum', 'Rain', 'Snow', 'TStorm'])
tweets = tweets.drop(tweets[tweets.tweet_dates < 50].index)
tweets.head()
```

* In order to merge the two dataframes we need to have a 'join' column, in this case the Date. It needs to be in the same data type. As we realised in the tweets data frame the date is in Float, we will convert it to integer.
* We will also change the name in the labels dataframe so it will be easier later.

```python
tweets['tweet_dates'] = tweets['tweet_dates'].fillna(0.0).astype(int)
```

```python
wlabels = wlabels.rename(columns = {'Day':'tweet_dates'})
dailyweather = dailyweather.rename(columns = {'Day':'tweet_dates'})

wlabels.head()
```

* Finally, we will use the merge function from pandas to join them based on the tweet_dates column

```python
merged1 = tweets.merge(wlabels, how='left', on='tweet_dates')
merged = merged1.merge(dailyweather, how='left', on='tweet_dates')
```

* And we will export it to csv.

```python
merged.to_csv('tweets_and_weather.csv')
```
